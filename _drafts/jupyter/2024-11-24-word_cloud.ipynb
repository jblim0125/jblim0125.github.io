## 웹 크롤링에 의한 데이터 수집과 저장

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
#Keyword 검색과 news_items 확보
keyword ='인공지능'
max_page = 2
news = ""
for p in range(max_page):
    url = f"https://search.hankookilbo.com/Search?Page={p+1}&tab=NEWS&sort=relation&searchText={keyword}&searchTypeSet=TITLE,CONTENTS&selectedPeriod=%EC%A0%84%EC%B2%B4&filter=head"
    resp = requests.get(url)
    html_src = resp.text
    soup = bs(html_src, 'html.parser')
    news_items = soup.select('div.img-box')
    # 각 뉴스 링크를 생성하고 기사를 종합
    for item in news_items:
        link = item.find('a').get("href")
        resp = requests.get(link)
        html_src = resp.text
        soup = bs(html_src, 'html.parser')
        # 기사에서 패러그래프 종합
        text=''
        for para in soup.select('p.editor-p'):
            text += para.text.strip()
        news +=text
        news +="\n\n"
news
```

```python
with open('인공지능.txt', 'w') as f:
    f.write(news)
```

## 텍스트 데이터 전처리

### 데이터 필터링
pandas를 이용해 csv파일을 로드하고 데이터를 확인한다.

```python
import pandas as pd
!wget 'https://raw.githubusercontent.com/byungjooyoo/Dataset/main/movie_comment.csv'
movie_data = pd.read_csv('movie_comment.csv')
movie_data= movie_data[['Title', 'Score','Comment']]
movie_data
```

제목이 인질이고, 점수가 10점인 데이터를 조회한다.

```python
con = (movie_data.Title == '인질') & (movie_data.Score == 10)
movie_data.loc[con, ['Score', 'Comment']]
```

제목 인질 데이터에서 마지막 10개와 데이터 제거, 컬럼 제거를 수행
점수가 1점과 6점인 데이터를 조회

```python
con_title = movie_data['Title']=='인질'
filter_data=movie_data.loc[con_title, : ].tail(10)
filter_data.drop(8121, axis=0, inplace= True)
#행 인덱스가 8121인 행을 삭제합니다.
#axis=0: 행을 기준으로 작업을 수행.
#inplace=True: 원본 데이터프레임(df1) 자체를 수정.
filter_data.drop("Title", axis = 1, inplace =True)
# df1 데이터프레임에서 열 이름이 'Title'인 열을 삭제합니다.
# axis=1: 열을 기준으로 작업을 수행.
# inplace=True: 원본 데이터프레임(df1) 자체를 수정.
con_score_1 = filter_data['Score'] == 1
con_score_6 = filter_data['Score'] == 6
filter_data.loc[con_score_1| con_score_6, : ]
```

### 결측치 및 중복 데이터 제거

데이터가 없는 행을 삭제한다.

```python
print(len(movie_data), movie_data.isnull().sum())
movie_data.dropna(inplace = True)
len(movie_data)
```

```python
data = movie_data.loc[movie_data['Title']=='인질', :]
data = data.reset_index()
data = data.loc[ : , ['Score', 'Comment']]
data
```

```python
print(data.duplicated().sum())
data.drop_duplicates(inplace=True)
data
```

### 데이터 탐색 과 이진데이터로 변환

```python
print(data['Score'].value_counts())
data['Score'].hist()
```

```python

data.loc[data['Score'] >= 6, 'Label'] = 1
data.loc[data['Score'] < 6, 'Label'] = 0
col = ['Comment','Label']
data.loc[:,col]
```

### 하나의 문서로 변환

```python
data1 = data.loc[data['Label'] == 1, ['Comment']]
data0 = data.loc[data['Label'] == 0, ['Comment']]
[x for x in data1['Comment']]
```

```python
text1 = ''
for para in data1['Comment']:
  text1 += para+'. '
print(text1)

text0 = []
for para in data0['Comment']:
  text0.append(para)
text0 = ".".join(text0)
print(text0)
```

## 탐색적 빈도 분석
### 형태소 분석기를 이용한 명사 추출

```python
import os

os.environ['JAVA_HOME'] = '/home/jblim/jdk-17.0.13+11'
os.environ['PATH'] = os.environ['JAVA_HOME'] + '/bin:' + os.environ['PATH']

!pip install konlpy
from konlpy.tag import Komoran

komoran = Komoran()
nouns1 = komoran.nouns(text1)
nouns0 = komoran.nouns(text0)
nouns0[:10]
```

### 데이터 정제와 빈도 추출

```python
nouns1 = [n for n in nouns1 if len(n) > 1]
nouns0 = [n for n in nouns0 if len(n) > 1]
```

```python
from collections import Counter
count1 = Counter(nouns1)
count0 = Counter(nouns0)
top1 = count1.most_common(100)
top0 = count0.most_common(100)
print(top1)
print(top0)
```

```python
!wget https://raw.githubusercontent.com/byungjooyoo/Dataset/main/korean_stopwords.txt
import re
stop_words=[]
with open("korean_stopwords.txt", 'r') as f:
    while True:
        line = f.readline().strip()
        stop_words.append(line)
        if not line: break
print(stop_words)
```

```python
stop_words = ['황정민', '영화']+stop_words
dic1 = dict(top1)
dic0 = dict(top0)
for word in stop_words:
    if word in dic1:
        del dic1[word]
    if word in dic0:
        del dic0[word]
print(dic1)
print(dic0)
```

```python
!pip install wordcloud

from wordcloud import WordCloud
from konlpy.tag import Komoran
import matplotlib.pyplot as plt
from konlpy.corpus import kolaw
# 대상 텍스트 데이터
kolaw.fileids()
para = kolaw.open('constitution.txt').read()
tagger = Komoran()
nouns = tagger.nouns(para)
text = ", ". join(nouns)
# 클라우딩 객체 설정
my_path='/content/NanumGothic.ttf'
wordcloud = WordCloud(font_path=my_path,
                      background_color='white',
                      width=1200,
                      height=800).generate(text)
fig = plt.figure(figsize=(10,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
```
